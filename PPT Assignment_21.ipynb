```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class DoublyLinkedListNode:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

def convert_to_doubly_linked_list(root):
    if root is None:
        return None

    # Initialize a dummy node for the doubly linked list
    dummy = DoublyLinkedListNode(None)

    # Convert the binary tree to doubly linked list
    prev = dummy
    convert_node(root, prev)

    # Adjust the head node of the doubly linked list
    head = dummy.next
    if head:
        head.prev = None

    return head

def convert_node(node, prev):
    if node is None:
        return prev

    # Convert the left subtree
    prev = convert_node(node.left, prev)

    # Convert the current node
    doubly_linked_node = DoublyLinkedListNode(node.value)
    prev.next = doubly_linked_node
    doubly_linked_node.prev = prev
    prev = doubly_linked_node

    # Convert the right subtree
    prev = convert_node(node.right, prev)

    return prev

# Example usage:
# Create a binary tree
root = TreeNode(4)
root.left = TreeNode(2)
root.right = TreeNode(6)
root.left.left = TreeNode(1)
root.left.right = TreeNode(3)
root.right.left = TreeNode(5)
root.right.right = TreeNode(7)

# Convert the binary tree to a doubly linked list
head = convert_to_doubly_linked_list(root)

# Traverse the doubly linked list forwards
current = head
while current:
    print(current.value, end=" ")
    current = current.next

# Output: 1 2 3 4 5 6 7

print()

# Traverse the doubly linked list backwards
current = head
while current and current.next:
    current = current.next
tail = current
while tail:
    print(tail.value, end=" ")
    tail = tail.prev

# Output: 7 6 5 4 3 2 1
```
